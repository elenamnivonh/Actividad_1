```python
#Importamos librerias requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Agregar el archivo: Inmuebles.xlsx

```python
#Cargamos desde un archivo .xlsx sin indice
df=pd.read_excel('Inmuebles.xlsx',skiprows=1)
df
```

Aplicar los siguientes filtros:


a) Eliminar la fila “0” de “Unnamed” y Eliminar la columna “0” y la columna “Referencia”


```python
filtroa= df.drop(['Referencia'], axis=1)
filtroa
```

b) Todos las registros de “Fecha Alta : igual a 2007-04-29 00:00:00 y a 2004-01-02 00:00:00


```python
filtrob=filtroa[(filtroa["Fecha_Alta"]=='2007-04-29 00:00:00')|(filtroa["Fecha_Alta"]=='2004-01-02 00:00:00')]
filtrob
```

c) Todos los registros de “ Tipo: parking, local, industrial y piso”


```python
filtroc=filtroa[(filtroa["Tipo"]=='Parking')|(filtroa["Tipo"]=='Local')|(filtroa["Tipo"]=='Industrial')|(filtroa["Tipo"]=='Piso')]
filtroc
```

d) Todos los registros con “Provincia” igual a “Girona” y con “Superficie” mayor a “100”


```python
filtrod=filtroa[(filtroa["Provincia"]=='Girona')& (filtroa["Superficie"]>100)]
filtrod
```

e) Todos los registros con “Precio Venta: mayor a 1000,000” o “vendedor” igual a “Carmen o Joaquin”


```python
filtroe=filtroa[(filtroa["Precio Venta"]>1000000)|(filtroa["Vendedor"]=="Carmen")|(filtroa["Vendedor"]=="Joaquín")]
filtroe
```

f) Todos los registros que no sean de “Provincia : Barcelona“


```python
filtrof=filtroa[~(filtroa["Provincia"]=='Barcelona')]
filtrof
```

g)Todos los registros que no sean de “Operación: alquiler “ ni “Tipo: oficina“



```python
filtrog=filtroa[~(filtroa["Operación"]=='Alquiler') & ~(filtroa["Tipo"]== 'Oficina')]
filtrog
```

Por cada filtro agregar un comentario en cada COMMIT y agregar el archivo .csv de cada filtro


```python
#CONVETIR LOS FILTROS A CSV
filtroa.to_csv("filtroa.csv")
filtrob.to_csv("filtrob.csv")
filtroc.to_csv("filtroc.csv")
filtrod.to_csv("filtrod.csv")
filtroe.to_csv("filtroe.csv")
filtrof.to_csv("filtrof.csv")
filtrog.to_csv("filtrog.csv")
```
