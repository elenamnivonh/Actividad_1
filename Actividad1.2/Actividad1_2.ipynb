```python
#Importamos librerias requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Cargamos desde un archivo .xlsx sin indice
df=pd.read_excel('Datos de Facturación.xlsx')
df
```

```python
#Verificamos la información del dataframe
df.info()
```

1-CVE_CLPV      (valores: 1000 a 2000)

```python
#Reemplazo de valores diferentes
df.CVE_CLPV=df.CVE_CLPV.replace({'8440Q':'8440'},regex=False)
```

```python
#Cambiamos el tipo de dato objeto por entero o flotante
df['CVE_CLPV']=df['CVE_CLPV'].astype(float)
```

```python
filtro121=df[(df["CVE_CLPV"]>=1000) &(df["CVE_CLPV"]<=2000)]
filtro121
```

2-CVE_VEND     (Todas las claves excepto “5” y “4”)

```python
filtro122=df[~(df["CVE_VEND"]==4) & ~(df["CVE_VEND"]== 5)]
filtro122
```

```python
#Comprobar que solo se omitieron las claves 4 y 5
prueba2=filtro122.groupby(['CVE_CLPV'])['CVE_CLPV'].count()
prueba2
```

3-FECHA_ENT   (Fechas de “28/02/2022”)

```python
df['FECHA_ENT']=df['FECHA_ENT'].astype(str)
filtro123=df[df['FECHA_ENT'] =='2022-04-13 00:00:00']
filtro123
```

4-CAN_TOT       (Cantidades menores a 5951.7)   o  STATUS (“E”)



```python
filtro124=df[(df["CAN_TOT"]<5951.7)| (df["STATUS"]=='E')]
filtro124
```

5-Solo las columnas: CVE_DOC, FECHA_ENT, FECHA_VEN y CAN_TOT



```python
filtro125=df.loc[:,["CVE_DOC","FECHA_ENT","FECHA_VEN","CAN_TOT"]]
filtro125
```

6-Solo las filas de 7001-7099

```python
filtro126=df.iloc[7001:7100]
filtro126
```

7-Index= CVE_VEND (valores: 1, 2)  (columna: FECHAELAB)

```python
df2 = pd.read_excel('Datos de Facturación.xlsx', index_col = 'CVE_VEND')
filtro127= df2.loc[[1,2], ['FECHAELAB']]
filtro127
```
