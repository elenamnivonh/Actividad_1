```python
#Importamos librerias requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Filtro básicos para variables

```python
#Filtro por objeto 
#Este filtro, asigna todas las ventas realizadas por un vendedor en específico
filtro1=df[df['NOMBRE_VENDEDOR'] =='LETICIA RAMIREZ HERNANDEZ']
filtro1
```

```python
#Filtro por filas 
#Se filtran las filas de la 10 a la 20
filtro2=df.iloc[10:21,:]
filtro2
```

```python
#Filtro por columnas
#Filtra el nombre del cliente y el costo total calculado
filtro3=df.iloc[ : ,[3,11]]
filtro3
```

```python
#Asignamos indice a CVE_DOC
df1=pd.read_excel("Detalle precios y productos fabricados 2022.xlsx",index_col=2)
df1
```

```python
#Filtra filas y columnas
#Filtramos al vendedor AARON NUÑO CORDOVA y las fechas de sus ventas
filtro4=df1.loc[["AARON NUÑO CORDOVA"],["FECHA_DOC"]]
filtro4
```

```python
#Filtro por cabecera
#Filtra los primeros 10 
filtro5=df.head(10)
filtro5
```

FILTRO BASIVOS PARA VARIABLES CUANTITATIVAS

```python
#Filtro por comparación 
#Filtra cuando el MARGEN_TOTAL_CALCULADO es mayor a 200
filtro6=df[df["MARGEN_TOTAL_CALCULADO"]>200]
filtro6
```

```python
#Filtro y
#Filtra cuando el margen total calculado es mayor a 200 y el precio unitario mayor o igual a 100
filtro7=df[(df["MARGEN_TOTAL_CALCULADO"]>200) &(df["PRECIO_UNITARIO"]>=100)]
filtro7
```

```python
#Filtro o
#Filtra cuando el marfent total calculado es menor a 100 o el precio unitario mayor a 200
filtro8=df[(df["MARGEN_TOTAL_CALCULADO"]<100)| (df["PRECIO_UNITARIO"]>200)]
filtro8
```

```python
#Filtro not 
#Filtro cuando el margen total calculado no es menor a 1000 y tampoco el precio unitario es igual a 100
filtro9=df[~(df["MARGEN_TOTAL_CALCULADO"]<1000) & ~(df["PRECIO_UNITARIO"]== 100)]
filtro9
```

```python
#CONVETIR LOS FILTROS A CSV
filtro1.to_csv("filtro1.csv")
filtro2.to_csv("filtro2.csv")
filtro3.to_csv("filtro3.csv")
filtro4.to_csv("filtro4.csv")
filtro5.to_csv("filtro5.csv")
filtro6.to_csv("filtro6.csv")
filtro7.to_csv("filtro7.csv")
filtro8.to_csv("filtro8.csv")
filtro9.to_csv("filtro9.csv")
```
